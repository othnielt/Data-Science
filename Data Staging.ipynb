## Transform

```python
# Cleaning the data types to coordinate it with what we want to do
# first we're going to remove all "Not Reported" , "Unknown" values since they're usefull and also focusing only on provinces no repatriated or territories

# discard rows only if all the conditions are met, you can use the all method with the conditions as a list comprehension.
df = df.loc[
    ~((df[['age_group', 'gender', 'exposure', 'case_status']] == 'Not Reported').all(axis=1))
]


#  Extracting only provinces like Alberta, Ontario, BC, Nova Scotia and Sasketchewan not territories and also to decrease our data since it's too large
provinces_to_keep = ['Alberta', 'British Columbia', 'Ontario', 'Quebec']
df = df[df['province'].isin(provinces_to_keep) & ~(df['province'].isin(['Repatriated', 'Nunavut', 'NWT', 'Yukon', 'Manitoba', 'New Brunswick', 'PEI', 'NL', 'Saskatchewan', 'Nova Scotia']))]




```

```python
# Dropping the unrequired columns
df = df.drop(['province_abbr','hr_uid', 'row_id', 'latitude', 'longitude'], axis=1)
```

## Generating measures/facts

```python
# create a groupby object by age group, gender, and health region
grouped_data = dataset_final.groupby(['age_group', 'gender', 'health_region'])

# Calculate the positivity rate and vaccination rate for each group
positivity_rates = grouped_data.apply(lambda x: x.loc[x['case_status']=='Active', 'Confirmed_Cases'].sum() / x['testing'].sum())
vaccination_rates = grouped_data.apply(lambda x: x['vaccine_administered'].sum() / total_pop_dataset * 100)

# Merge the results back into the original dataframe
dataset_final = dataset_final.merge(positivity_rates.rename('Test_positivity_rate'), on=['age_group', 'gender', 'health_region'])
dataset_final = dataset_final.merge(vaccination_rates.rename('vaccination_rate'), on=['age_group', 'gender', 'health_region'])

# calculate Total_fatalities and Total_recorded_cases for each group
grouped_totals = grouped_data.agg({'Confirmed_Cases': 'sum', 'Deaths': 'sum'})

# calculate Total_fatalities and Total_recorded_cases for the entire dataset
dataset_final['Total_fatalities'] = grouped_totals['Deaths'].sum()
dataset_final['Total_recorded_cases'] = grouped_totals['Confirmed_Cases'].sum()
```

## Generating Surrogate Keys

```python
dataset_final['Surrogate Keys'] = range(1,len(dataset_final)+1)
```

```python
dataset_final
```

```python
dataset = dataset_final.reindex(columns=['Surrogate Keys'] + list([c for c in dataset_final.columns if c!= 'Surrogate Keys']))
```
